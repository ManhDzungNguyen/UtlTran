```python
import json
import pandas as pd
```

```python
file_path_train = '/content/drive/MyDrive/Colab Notebooks/MRC - VLSP/Data/train_xquad.json'
file_path_dev = '/content/drive/MyDrive/Colab Notebooks/MRC - VLSP/Data/dev_xquad.json'
file_path_test = '/content/drive/MyDrive/Colab Notebooks/MRC - VLSP/Data/test_xquad.json'
```

```python
# Opening JSON file
with open(file_path_train) as json_file:
  data_train = json.load(json_file)

with open(file_path_dev) as json_file:
  data_dev = json.load(json_file)

with open(file_path_test) as json_file:
  data_test = json.load(json_file)
```

```python
# data.keys() # 'version', 'data'
# type(data['data']) # list
# data['data'][0].keys() # 'title', 'paragraphs'
# data['data'][0]['paragraphs'][0].keys() # 'context', 'qas'
# data['data'][0]['paragraphs'][0]['context']
# data['data'][0]['paragraphs'][0]['qas'][0].keys() # 'id', 'question', 'answers'
# data['data'][0]['paragraphs'][0]['qas'][0]['id']
# data['data'][0]['paragraphs'][0]['qas'][0]['question']
# data['data'][0]['paragraphs'][0]['context']
# type(data['data'][0]['paragraphs'][0]['qas'][0]['question'])
```

```python
ls_train = []
for paragraphs in data_train['data']:
  for context in paragraphs['paragraphs']:
    for qas in context['qas']:
      for answer in qas['answers']:
        dic = {}
        dic['context'] = context['context']
        dic['question'] = qas['question']
        dic['answer'] = answer['text']
        dic['title'] = paragraphs['title']
        ls_train.append(dic)

df_train = pd.DataFrame(ls_train)
len(df_train)
```

```python
ls_dev = []
for paragraphs in data_dev['data']:
  for context in paragraphs['paragraphs']:
    for qas in context['qas']:
      for answer in qas['answers']:
        dic = {}
        dic['context'] = context['context']
        dic['question'] = qas['question']
        dic['answer'] = answer['text']
        dic['title'] = paragraphs['title']
        ls_dev.append(dic)

df_dev = pd.DataFrame(ls_dev)
len(df_dev)
```

```python
ls_dev = []
for paragraphs in data_test['data']:
  for context in paragraphs['paragraphs']:
    for qas in context['qas']:
      for answer in qas['answers']:
        dic = {}
        dic['context'] = context['context']
        dic['question'] = qas['question']
        dic['answer'] = answer['text']
        dic['title'] = paragraphs['title']
        ls_dev.append(dic)

df_dev = pd.DataFrame(ls_dev)
len(df_dev)
```

```python
df_all = pd.concat([df_train, df_dev, df_test])
df_all.describe()
```
