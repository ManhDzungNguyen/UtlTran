```
import json
import pandas as pd
import numpy as np
from tqdm import tqdm
import random
```

```
raw_dataset = '/content/drive/MyDrive/Colab Notebooks/MRC - VLSP/Dataset/train_data.json'
```

```
with open(raw_dataset) as raw_file:
  raw_dataset_v2 = json.load(raw_file)
```

```
ls_raw_dataset_v2 = []
for paragraphs in raw_dataset_v2['data']:
  for context in paragraphs['paragraphs']:
    for qas in context['qas']:
      dic = {}
      dic['title'] = paragraphs['title']
      dic['context'] = context['context']
      dic['id'] = qas['id']
      dic['question'] = qas['question']
      dic['is_impossible'] = qas['is_impossible']
      dic['answers'] = qas['answers']
      dic['plausible_answers'] = []
      if dic['is_impossible']:
        dic['plausible_answers'] = qas['plausible_answers']
      

      ls_raw_dataset_v2.append(dic)
```

# No-answer generotor

```
ls_raw_dataset_v2_1 = ls_raw_dataset_v2.copy()
index = 28458
cnt = 0
```

```
import random

ls_no_answer_genarate = []

while True:
  random_index_1 = random.randint(0,len(ls_raw_dataset_v2_1)-1)
  row_1 = ls_raw_dataset_v2_1.pop(random_index_1)
  title = row_1['title']

  random_index_2 = random.randint(0,len(ls_raw_dataset_v2_1)-1)
  row_2 = ls_raw_dataset_v2_1[random_index_2]

  while(row_2['title'] == title):
    random_index_2 = random.randint(0,len(ls_raw_dataset_v2_1)-1)
    row_2 = ls_raw_dataset_v2_1[random_index_2]

  add_dic = {}
  add_dic['title'] = row_1['title']
  add_dic['context'] = row_1['context']
  add_dic['question'] = row_2['question']
  add_dic['plausible_answers'] = []
  add_dic['answers']=[]
  add_dic['id'] = 'uit_0'+str(index)
  add_dic['is_impossible'] = True
  index = index + 1
  cnt = cnt + 1
  ls_no_answer_genarate.append(add_dic)
  if cnt == 7016: 
    break

  
len(ls_no_answer_genarate)
```

# Export dataset to JSON file

```
structure_squadv2 = {
    'version' : 'viquad2_training_set',
    'data' : []
}
```

```
for idx, row in tqdm(df_final_dataset.iterrows()):
  if row['is_impossible']:
    sample = {
      'paragraphs': [
        {
          'context': row['context'],
          'qas': [
            { 
              'answers': row['answers'],
              'id': row['id'],
              'question': row['question'],
              'is_impossible': row['is_impossible'],
              'plausible_answers': row['plausible_answers']       
            }
          ]
        }
      ],
      'title': row['title']
    }
    structure_squadv2['data'].append(sample)

  else:
    sample = {
      'paragraphs': [
        {
          'context': row['context'],
          'qas': [
            { 
              'answers': row['answers'],
              'id': row['id'],
              'question': row['question'],
              'is_impossible': row['is_impossible'],      
            }
          ]
        }
      ],
      'title': row['title']
    }
    structure_squadv2['data'].append(sample)
```
