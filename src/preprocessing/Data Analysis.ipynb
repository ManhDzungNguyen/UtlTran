```python
Corpus = pd.read_csv(r'/content/drive/MyDrive/Colab Notebooks/Question Classification/Data/finalSampling_train_03.csv',encoding='utf-8')
```

```python
print(Corpus.groupby(['label']).count())
```

```python
def sampling_data(data, column, sampling_type = 'oversampling'):
  new_data = data
  unique_values = data[column].unique()
  df = data[column].value_counts()
  
  if sampling_type == 'oversampling':
    num_of_label = df.max()
    for label in unique_values:
      add = num_of_label - df[label]
      for i in range(add):
        sub_data = data.loc[data[column] == label].sample(n = 1)
        new_data = pd.concat([new_data, sub_data])

  elif sampling_type == 'undersampling':
    num_of_label = df.min()
    for label in unique_values:
      subtract = df[label] - num_of_label
      for i in range(subtract):
        new_data = new_data.drop(new_data.loc[new_data[column] == label].sample(n = 1).index)

  elif sampling_type == 'averagesampling':
    num_of_label = int(df.mean())
    for label in unique_values:
      change = num_of_label - df[label]
      if change > 0:
        for i in range(change):
          sub_data = data.loc[data[column] == label].sample(n = 1)
          new_data = pd.concat([new_data, sub_data])
      elif change <0:
        for i in range(abs(change)):
          new_data = new_data.drop(new_data.loc[new_data[column] == label].sample(n = 1).index)
    
  else:
    try:
      num_of_label = int(sampling_type)
      if num_of_label > 0:
        for label in unique_values:
          change = num_of_label - df[label]
          if change > 0:
            for i in range(change):
              sub_data = data.loc[data[column] == label].sample(n = 1)
              new_data = pd.concat([new_data, sub_data])
          elif change <0:
            for i in range(abs(change)):
                new_data = new_data.drop(new_data.loc[new_data[column] == label].sample(n = 1).index)
      else:
        print('sampling_type: invalid value!')

    except:
      print('sampling_type: invalid value!')

  return new_data
```

```python
New_Corpus = sampling_data(Corpus, 'label', sampling_type = 400)
print(New_Corpus.groupby(['label']).count())
```
